```python
from abc import *
from accessify import protected, private

#абстрактний метаклас Data_Storage
class Data_Storage(metaclass = ABCMeta):
    
    #конструктор
    def __init__(self, capacity, reserved_storage=0):
        self.capacity = capacity
        self.reserved_storage = reserved_storage
    
    #прихований абстрактний метод clear_storage
    @protected
    @abstractmethod
    def clear_storage(self):
        self.reserved_storage = 0
        print('Storage cleared\nCapacity: {0}\nReserved Storage: {1}'.format(self.capacity, self.reserved_storage))
        
    #прихований абстрактний метод clear_storage    
    @protected
    @abstractmethod
    def add_item(self, size_of_item):
        if self.capacity - self.reserved_storage > size_of_item:
            self.reserved_storage += size_of_item
            print('Item added\nCapacity: {0}\nReserved Storage: {1}'.format(self.capacity, self.reserved_storage))
        else:
            print('Storage is full\nCapacity: {0}\nReserved Storage: {1}\nSize of item: {2}'.format(self.capacity, self.reserved_storage, size_of_item))
    

#клас Web_storage, що наслідує Data_Storage
class Web_storage(Data_Storage):
    
    #конструктор
    def __init__(self, capacity, reserved_storage = 0):
        if 0 < capacity <= 1:
            Data_Storage.__init__(self, capacity, reserved_storage)
            print('Created Web Storage\nCapacity: {0}\nReserved Storage: {1}'.format(self.capacity, self.reserved_storage))
        else:
            raise ValueError("Capacity should be greater than 0 and lesser than 1. Received capacity: {}.".format(self.capacity))
            
    #віртуальні функції clear_storage і add_item
    def clear_storage(self):
        Data_Storage.clear_storage(self)
        
    def add_item(self, size_of_item):
        Data_Storage.add_item(self, size_of_item)

        
        
#клас Removable_storage, що наслідує Data_Storage
class Removable_storage(Data_Storage):
    
    #конструктор
    def __init__(self, capacity, reserved_storage = 0):
        if 1 < capacity <= 64:
            Data_Storage.__init__(self, capacity, reserved_storage)
            print('Created Removable Storage\nCapacity: {0}\nReserved Storage: {1}'.format(self.capacity, self.reserved_storage))
        else:
            raise ValueError("Capacity should be greater than 1 and lesser than 64. Received capacity: {}.".format(capacity))
            
    #віртуальні функції clear_storage і add_item
    def clear_storage(self):
        Data_Storage.clear_storage(self)
        
    def add_item(self, size_of_item):
        Data_Storage.add_item(self, size_of_item)
        
        
        
#клас Local_storage, що наслідує Data_Storage
class Local_storage(Data_Storage):
    
    #констуктор
    def __init__(self, capacity, reserved_storage = 0):
        if capacity > 64:
            Data_Storage.__init__(self, capacity, reserved_storage)
            print('Created Local Storage\nCapacity: {0}\nReserved Storage: {1}'.format(self.capacity, self.reserved_storage))
        else:
            raise ValueError("Capacity should be greater than 64. Received capacity: {}.".format(capacity))
       
    #віртуальні функції clear_storage і add_item
    def clear_storage(self):
        Data_Storage.clear_storage(self)
        
    def add_item(self, size_of_item):
        Data_Storage.add_item(self, size_of_item)
        
    #абстрактний метод change_speed    
    @abstractmethod
    def change_speed(self, speed):
        self.speed = speed
        
    #захищена функція add_capacity    
    @private
    def add_capacity(self, capacity):
        self.capacity += capacity

        

#клас SSD_Local_storage, що наслідує Local_storage        
class SSD_Local_storage(Local_storage):
    
    #конструктор
    def __init__(self, capacity, reserved_storage=0):
        print('SSD')
        Local_storage.__init__(self, capacity, reserved_storage)
        self.speed = 400
        
    #віртуальна функція change_speed
    def change_speed(self, speed):
        Local_storage.change_speed(self, speed)
        
        

#клас HDD_Local_storage, що наслідує Local_storage                
class HDD_Local_storage(Local_storage):
    
    #конструктор
    def __init__(self, capacity, reserved_storage=0):
        print('HDD')
        Local_storage.__init__(self, capacity, reserved_storage)
        self.speed = 100
    
    #віртуальна функція change_speed
    def change_speed(self, speed):
        Local_storage.change_speed(self, speed)
    
```
